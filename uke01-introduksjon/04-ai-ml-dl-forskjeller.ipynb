## üî¨ Maskinl√¶ring: Fra mystikk til matematikk

### Hvorfor matematikk? ü§î

Mange synes maskinl√¶ring virker mystisk og komplisert. **Men i bunn og grunn er det bare matematikk!** 

N√•r vi beskriver maskinl√¶ring matematisk, oppn√•r vi tre viktige ting:

1. **Klarhet** üéØ - Vi forst√•r presist hva som menes og hva som skjer
2. **Kommunikasjon** üí¨ - Vi kan snakke presist med kollegaer og forskere
3. **Probleml√∏sning** üõ†Ô∏è - Vi kan feils√∏ke og forbedre systemene v√•re

### Tenk p√• det slik...

I medisin bruker vi ogs√• matematikk for √• beskrive komplekse prosesser:
- **Blodtrykk**: 120/80 mmHg
- **Medisindosering**: mg/kg kroppsvekt  
- **Laboratorieverdier**: Referanseomr√•der og enheter

P√• samme m√•te gir matematikk oss et **presist spr√•k** for √• beskrive hvordan AI-systemer fungerer.

### Den gode nyheten üìà

Du trenger **ikke** v√¶re matematiker for √• forst√• grunnprinsippene! 

Akkurat som du ikke trenger √• v√¶re kjemiker for √• forst√• at insulin senker blodsukkeret, trenger du ikke v√¶re matematiker for √• forst√• at maskinl√¶ring l√¶rer m√∏nstre fra data.

La oss se p√• den **enkle matematiske beskrivelsen** av maskinl√¶ring:

```python
# Visualiser maskinl√¶ring-prosessen
fig, ax = plt.subplots(figsize=(16, 10))

# Sett bakgrunnsfarge til lys
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# Definer posisjoner for boksene
input_pos = (0.15, 0.5)
algo_pos = (0.5, 0.5)
output_pos = (0.85, 0.5)

# Tegn hovedboksene
input_box = plt.Rectangle((input_pos[0]-0.08, input_pos[1]-0.15), 0.16, 0.3, 
                         facecolor='lightgray', edgecolor='black', linewidth=2)
ax.add_patch(input_box)

algo_box = plt.Rectangle((algo_pos[0]-0.12, algo_pos[1]-0.15), 0.24, 0.3, 
                        facecolor='lightgray', edgecolor='black', linewidth=2)
ax.add_patch(algo_box)

output_box = plt.Rectangle((output_pos[0]-0.08, output_pos[1]-0.15), 0.16, 0.3, 
                          facecolor='lightgray', edgecolor='black', linewidth=2)
ax.add_patch(output_box)

# Legg til tekst i boksene
ax.text(input_pos[0], input_pos[1], 'Input', ha='center', va='center', 
        fontsize=30, fontweight='bold', color='black')

ax.text(algo_pos[0], algo_pos[1], 'Algoritme', ha='center', va='center', 
        fontsize=30, fontweight='bold', color='black')

ax.text(output_pos[0], output_pos[1], 'Output', ha='center', va='center', 
        fontsize=30, fontweight='bold', color='black')

# Tegn piler mellom boksene
arrow1 = plt.Arrow(input_pos[0]+0.09, input_pos[1], 0.15, 0, 
                   width=0.08, color='gray', alpha=0.8)
ax.add_patch(arrow1)

arrow2 = plt.Arrow(algo_pos[0]+0.13, algo_pos[1], 0.15, 0, 
                   width=0.08, color='gray', alpha=0.8)
ax.add_patch(arrow2)

# Legg til matematisk notasjon √∏verst
ax.text(0.5, 0.85, r'$y \approx f(x; \theta)$', ha='center', va='center', 
        fontsize=70, style='italic', color='black', fontweight='bold')

# Input forklaring
ax.text(input_pos[0], 0.25, 'x', ha='center', va='top', 
        fontsize=50, style='italic', color='black', fontweight='bold')
ax.text(input_pos[0], 0.15, 'data', ha='center', va='top', 
        fontsize=40, fontweight='bold', color='black')
input_details = 'tabell,\ntekst,\nbilde,\nlyd, ...'
ax.text(input_pos[0], 0.02, input_details, ha='center', va='top', 
        fontsize=30, color='black')

# Output forklaring
ax.text(output_pos[0], 0.25, 'y', ha='center', va='top', 
        fontsize=50, style='italic', color='black', fontweight='bold')
ax.text(output_pos[0], 0.15, 'utfall', ha='center', va='top', 
        fontsize=40, fontweight='bold', color='black')
output_details = 'diagnose,\nprognose,\ngrad, type,\nklasse,\nvarighet, ...'
ax.text(output_pos[0], 0.02, output_details, ha='center', va='top', 
        fontsize=30, color='black')

# Algoritme forklaring - plasser f(x; theta) f√∏rst
formula_pos = (algo_pos[0], 0.25)
ax.text(formula_pos[0], formula_pos[1], r'$f(x; \theta)$', ha='center', va='top', 
        fontsize=56, style='italic', color='black', fontweight='bold')

# Separate forklaringer til venstre og h√∏yre med piler som peker p√• riktige deler
# Venstre side - peker p√• f
ax.text(algo_pos[0]-0.15, 0.15, 'modell', ha='center', va='top', 
        fontsize=40, fontweight='bold', color='red')
ax.text(algo_pos[0]-0.15, 0.05, '(funksjon)', ha='center', va='top', 
        fontsize=32, color='red')

# H√∏yre side - peker p√• theta
ax.text(algo_pos[0]+0.15, 0.15, 'modell-', ha='center', va='top', 
        fontsize=38, fontweight='bold', color='blue')
ax.text(algo_pos[0]+0.15, 0.08, 'parametre', ha='center', va='top', 
        fontsize=38, fontweight='bold', color='blue')
ax.text(algo_pos[0]+0.15, -0.02, '(l√¶res fra data)', ha='center', va='top', 
        fontsize=32, color='blue')

# Piler som peker p√• riktige deler av formelen
# Pil til f (venstre del av formelen)
ax.annotate('', xy=(algo_pos[0]-0.08, formula_pos[1]-0.02), 
            xytext=(algo_pos[0]-0.15, 0.16),
            arrowprops=dict(arrowstyle='->', color='red', lw=4))

# Pil til theta (h√∏yre del av formelen) 
ax.annotate('', xy=(algo_pos[0]+0.06, formula_pos[1]-0.02), 
            xytext=(algo_pos[0]+0.15, 0.16),
            arrowprops=dict(arrowstyle='->', color='blue', lw=4))

# Sett aksegrenser og fjern akser
ax.set_xlim(0, 1)
ax.set_ylim(-0.1, 1)
ax.axis('off')

plt.tight_layout()
plt.show()

print("üî¨ Maskinl√¶ring: Input (x) ‚Üí Algoritme f(x;Œ∏) ‚Üí Output (y)")
print("üìä Algoritmen l√¶rer parametere (Œ∏) fra treningsdata for √• lage prediksjoner")
print("üéØ R√∏d pil peker p√• f (modell/funksjon), bl√• pil peker p√• Œ∏ (parametere)")
print("\n VIKTIG: Symbolet '‚âà' betyr 'tiln√¶rmet lik' eller 'approksimerer'")
print("üéØ Hvorfor ikke '=' (likhetstegn)?")
print("   ‚Ä¢ Modeller gir tiln√¶rminger, ikke perfekte svar")
print("   ‚Ä¢ Det er ‚âàalltid usikkerhet og feil i prediksjoner")
print("   ‚Ä¢ I helse: En AI kan si '85% sannsynlighet for diabetes', ikke 'definitivt diabetes'")
print("\nüí° Dette minner oss p√• at AI-resultater alltid m√• tolkes med forsiktighet!")
```

## üéØ Tre viktige begreper: 
### **Persontilpasset-** vs **Presisjons-** vs **P4-medisin**

### üéØ **Presisjonsmedisin (Precision Medicine)**
- **Fokus**: H√∏y n√∏yaktighet og spesifisitet i diagnostikk og behandling
- **Tiln√¶rming**: "Riktig tiltak til riktig person p√• riktig tidspunkt med riktig metode"
- **Bredere enn gener**: 
  - üß¨ **Presisjonsgenetikk**: Kreftbehandling basert p√• tumorens DNA
  - üñºÔ∏è **Presisjonsavbildning**: AI-analyse av MRI for √• oppdage tidlige forandringer
  - üß† **Presisjonspsykologi**: M√•lrettet kognitiv terapi basert p√• nevroprofil
  - üíä **Presisjonsfarmakologi**: Medikamentvalg og -dosering basert p√• metaboliske profiler
- **Tidshorisont**: N√•r presis diagnose/behandling er n√∏dvendig

### üë§ **Persontilpasset medisin (Personalized Medicine)**  
- **Fokus**: Individuelle karakteristika (biologi + milj√∏ + preferanser)
- **Tiln√¶rming**: "Skredders√∏m" for hver enkelt pasient som helhetlig person
- **Eksempel**: Behandlingsplan som tar hensyn til pasientens verdier, familie, og kanskje √∏konomi
- **Tidshorisont**: Behandling tilpasset individets unike situasjon

### üîÑ **P4-medisin (P4 Medicine)**
- **Fokus**: Fire P-er: **Predictive, Preventive, Personalized, Participatory**
- **Tiln√¶rming**: Helhetlig, fremtidsrettet helsetjeneste-√∏kosystem
- **Eksempel**: AI forutsier diabetes-risiko ‚Üí forebyggende tiltak ‚Üí individuell plan ‚Üí pasient deltar aktivt
- **Tidshorisont**: Fra f√∏dsel til d√∏d (hele livsl√∏pet)

### üìä **Sammenligning**:

| *Aspekt* | Presisjon | Persontilpasset | P4-medisin |
|--------|-----------|---------------|-----|
| **Hovedfokus** | H√∏y n√∏yaktighet | Individuell tilpasning | Helhetlig system |
| **Metode** | Avansert teknologi | Skredders√∏m | Systemisk tiln√¶rming |
| **Omfatter** | Spesifikke tilstander | Hele personen | Hele populasjonen |
| **Timing** | Ved diagnosebehov | Ved behandling | Hele livet |
| **Teknologi** | AI, avanserte sensorer | Multimodale data | √òkosystem av verkt√∏y |
| **Pasientrolle** | M√•lpunkt | Partner | Aktiv deltaker |

### üîó **Hvordan de overlapper**:
- **Presisjonsmedisin** gir oss **verkt√∏yene** (n√∏yaktige m√•linger, AI-diagnoser)
- **Persontilpasset medisin** bruker verkt√∏yene for **individuell tilpasning**  
- **P4-medisin** setter alt inn i et **forebyggende, deltakende system**

### ü§ñ **AI's rolle**:
- **Presisjon**: AI gir h√∏y (dels overmenneskelig) n√∏yaktighet i analyse
- **Personalisering**: AI kombinerer komplekse datatyper for individuell tilpasning
- **P4**: AI gj√∏r prediksjoner mulig og st√∏tter deltakelse

### üîó **Forbindelse til v√•r figur**:

**üí° Enkelt sagt**:

## üß† Maskinl√¶ring (ML) - AI som l√¶rer

### Definisjon
**Maskinl√¶ring er en delmengde av AI hvor systemer l√¶rer og forbedrer seg automatisk fra erfaring (data) uten √• v√¶re eksplisitt programmert.**

### Kjennetegn
- L√¶rer fra data
- Forbedrer prestasjon over tid
- Generaliserer til nye situasjoner
- Oppdager m√∏nstre autonomt

```python
# ML-typer og deres egenskaper
ml_typer = pd.DataFrame({
    'Type': ['Supervised Learning', 'Unsupervised Learning', 'Reinforcement Learning'],
    'Databehov': ['Merkede data (input + output)', 'Umerkede data (kun input)', 'Milj√∏ med bel√∏nning/straff'],
    'M√•l': ['Predikere utfall', 'Finne skjulte m√∏nstre', 'Optimalisere handlinger'],
    'Helse_eksempel': [
        'Predikere hjerteinfarkt fra EKG',
        'Gruppere pasienter med like symptomer',
        'Optimalisere medikamentdosering'
    ]
})

print("üéì De tre hovedtypene av maskinl√¶ring:")
display(ml_typer)
```

```python
# Visualiser ML-l√¶ringsprosessen
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Supervised Learning
axes[0].scatter([1, 2, 3, 4], [2, 4, 6, 8], color='blue', s=100, alpha=0.7)
axes[0].plot([1, 4], [2, 8], 'r--', linewidth=2, label='L√¶rd modell')
axes[0].set_title('Supervised Learning\n(L√¶r fra eksempler)', fontweight='bold')
axes[0].set_xlabel('Input (f.eks. alder)')
axes[0].set_ylabel('Output (f.eks. risiko)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Unsupervised Learning
np.random.seed(42)
group1 = np.random.normal([2, 2], 0.5, (20, 2))
group2 = np.random.normal([6, 6], 0.5, (20, 2))
axes[1].scatter(group1[:, 0], group1[:, 1], color='red', alpha=0.7, s=50)
axes[1].scatter(group2[:, 0], group2[:, 1], color='blue', alpha=0.7, s=50)
axes[1].set_title('Unsupervised Learning\n(Finn m√∏nstre)', fontweight='bold')
axes[1].set_xlabel('Egenskap 1')
axes[1].set_ylabel('Egenskap 2')
axes[1].grid(True, alpha=0.3)

# Reinforcement Learning
states = range(5)
rewards = [1, 3, 5, 8, 10]
axes[2].bar(states, rewards, color='green', alpha=0.7)
axes[2].set_title('Reinforcement Learning\n(L√¶r fra bel√∏nning)', fontweight='bold')
axes[2].set_xlabel('Handling')
axes[2].set_ylabel('Bel√∏nning')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("üîç Hver ML-type l√∏ser forskjellige typer problemer!")
```

## üï∏Ô∏è Dypl√¶ring (DL) - ML med nevrale nettverk

### Definisjon
**Dypl√¶ring er en delmengde av maskinl√¶ring som bruker nevrale nettverk med mange lag ("dype" nettverk) for √• l√¶re komplekse m√∏nstre.**

### Kjennetegn
- Nevrale nettverk med mange lag
- Automatisk feature engineering
- H√•ndterer komplekse data (bilder, tekst, lyd)
- Krever mye data og regnekraft

```python
# DL-arkitekturer og deres bruksomr√•der
dl_arkitekturer = pd.DataFrame({
    'Arkitektur': ['CNN\n(Convolutional)', 'RNN\n(Recurrent)', 'Transformer', 'GAN\n(Generative)'],
    'Best_for': ['Bilder', 'Sekvenser', 'Spr√•k', 'Generering'],
    'Helse_eksempel': [
        'R√∏ntgen-analyse',
        'EKG-tolkning',
        'Medisinsk rapportskriving',
        'Syntetiske medisinske bilder'
    ],
    'Kompleksitet': [7, 6, 9, 8],
    'Data_behov': ['H√∏y', 'Middels', 'Veldig h√∏y', 'H√∏y']
})

print("üèóÔ∏è Dypl√¶ring-arkitekturer og deres spesialiteter:")
display(dl_arkitekturer)
```

```python
# Visualiser et enkelt nevral nettverk
fig, ax = plt.subplots(figsize=(12, 8))

# Definer lag-posisjoner
layer_positions = [1, 3, 5, 7]
layer_sizes = [3, 5, 4, 2]
layer_names = ['Input\n(Pasient-data)', 'Skjult lag 1\n(Feature detection)', 
               'Skjult lag 2\n(Pattern recognition)', 'Output\n(Diagnose)']

# Tegn noder
for i, (x, size, name) in enumerate(zip(layer_positions, layer_sizes, layer_names)):
    y_positions = np.linspace(1, 7, size)
    
    for y in y_positions:
        circle = Circle((x, y), 0.3, fill=True, facecolor='lightblue', 
                       edgecolor='blue', alpha=0.7)
        ax.add_patch(circle)
    
    # Legg til lag-navn
    ax.text(x, 0.2, name, ha='center', va='center', fontsize=10, 
            fontweight='bold', rotation=0)

# Tegn forbindelser mellom lag
for i in range(len(layer_positions)-1):
    x1, x2 = layer_positions[i], layer_positions[i+1]
    size1, size2 = layer_sizes[i], layer_sizes[i+1]
    
    y1_positions = np.linspace(1, 7, size1)
    y2_positions = np.linspace(1, 7, size2)
    
    for y1 in y1_positions:
        for y2 in y2_positions:
            ax.plot([x1+0.3, x2-0.3], [y1, y2], 'gray', alpha=0.7, linewidth=0.5)

ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title('Dypl√¶ring: Nevral nettverks-arkitektur', fontsize=16, fontweight='bold', pad=20)

# Legg til forklarende tekst
ax.text(4, 7.75, 'Informasjon flyter fra venstre til h√∏yre\nHvert lag l√¶rer mer komplekse m√∏nstre', 
        ha='center', va='center', fontsize=12, style='italic',
        bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.3))

plt.tight_layout()
plt.show()

print("üß† Hvert lag i nettverket l√¶rer √• gjenkjenne mer og mer komplekse m√∏nstre!")
```
